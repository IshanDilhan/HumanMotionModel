# Step 1: Dataset Preparation

Download JHMDB → Extract → Process for training

## 1. Setup & Imports

```python
import os
import cv2
import numpy as np
import mediapipe as mp
from pathlib import Path
from tqdm import tqdm
import shutil

print("✓ Imports successful")

# Set working directory
os.chdir(os.getcwd())
print(f"Working directory: {os.getcwd()}")
```

## 2. Verify JHMDB Dataset Structure

```python
DATA_DIR = Path("data/JHMDB")

print("\n" + "="*60)
print("JHMDB Dataset Setup")
print("="*60)

if not DATA_DIR.exists():
    print(f"\n❌ JHMDB dataset NOT found at {DATA_DIR}")
    print("\n📥 Manual Download Steps:")
    print("1. Visit: http://jhmdb.org/")
    print("2. Download 'Renamed Folders' (~348 MB)")
    print(f"3. Extract to: {DATA_DIR}/")
    print("4. Run this cell again\n")
else:
    classes = sorted([d.name for d in DATA_DIR.iterdir() if d.is_dir()])
    print(f"\n✓ Found {len(classes)} action classes")
    print("\nClasses:")
    for cls in classes[:20]:
        cls_dir = DATA_DIR / cls
        videos = list(cls_dir.glob("*.avi")) + list(cls_dir.glob("*.mp4"))
        print(f"  {cls:25s} - {len(videos):3d} videos")
    if len(classes) > 20:
        print(f"  ... and {len(classes) - 20} more")
```

## 3. Map JHMDB Actions to Your 9 Motion Classes

```python
# Motion class mapping
MOTION_LABELS = [
    "Stationary",       # 0
    "Approaching",      # 1
    "Backing Away",     # 2
    "Across",           # 3
    "Slow Approach",    # 4
    "Fast Toward",      # 5
    "Fast Across",      # 6
    "Approach + Stop",  # 7
    "Minimal",          # 8
]

# Map JHMDB actions to motion labels
ACTION_TO_MOTION = {
    # Approaching (moving toward camera)
    'running': 1,
    'walking': 1,
    'climbing': 1,
    'jumping': 1,
    'pushing': 1,
    'pulling': 1,
    
    # Backing Away (moving away from camera)
    'falling': 2,
    'falling_down': 2,
    
    # Across (lateral movement)
    'waving': 3,
    'brushing_teeth': 3,
    'combing_hair': 3,
    
    # Stationary
    'sitting_down': 0,
    'standing': 0,
    'sit': 0,
    
    # Slow Approach
    'crawling': 4,
    'typing': 4,
    
    # Fast Toward
    'kicking': 5,
    'punching': 5,
    
    # Fast Across
    'shaking_hands': 6,
    'clapping': 6,
    
    # Approach + Stop
    'jogging': 7,
    
    # Minimal
    'reading': 8,
    'writing': 8,
    'looking_around': 8,
}

print("Motion Class Mapping:")
print("=" * 60)
for i, label in enumerate(MOTION_LABELS):
    actions = [a for a, m in ACTION_TO_MOTION.items() if m == i]
    print(f"\n{i}: {label:20s}")
    print(f"   Actions: {', '.join(actions)}")
```

## 4. Organize Videos by Motion Type

```python
SELECTED_MOTIONS_DIR = Path("data/selected_motions")
SELECTED_MOTIONS_DIR.mkdir(parents=True, exist_ok=True)

# Create directories for each motion type
for i, motion_label in enumerate(MOTION_LABELS):
    motion_dir = SELECTED_MOTIONS_DIR / f"{i:02d}_{motion_label.lower().replace(' ', '_')}"
    motion_dir.mkdir(exist_ok=True)

print(f"✓ Created directories in {SELECTED_MOTIONS_DIR}")

# Copy/symlink videos from JHMDB
video_counts = {i: 0 for i in range(len(MOTION_LABELS))}

print("\nOrganizing videos...")
for jhmdb_action in tqdm(DATA_DIR.iterdir()):
    if not jhmdb_action.is_dir():
        continue
    
    action_name = jhmdb_action.name.lower()
    motion_id = ACTION_TO_MOTION.get(action_name, None)
    
    if motion_id is None:
        continue
    
    # Get all videos in this action folder
    videos = list(jhmdb_action.glob("*.avi")) + list(jhmdb_action.glob("*.mp4"))
    
    motion_dir = SELECTED_MOTIONS_DIR / f"{motion_id:02d}_{MOTION_LABELS[motion_id].lower().replace(' ', '_')}"
    
    for video in videos:
        dest = motion_dir / video.name
        if not dest.exists():
            # Create symlink (efficient, doesn't copy data)
            try:
                os.symlink(video, dest)
                video_counts[motion_id] += 1
            except:
                # Fallback to copy if symlink fails
                shutil.copy2(video, dest)
                video_counts[motion_id] += 1

print("\n✓ Videos organized by motion type:")
print("=" * 60)
total_videos = 0
for i, motion_label in enumerate(MOTION_LABELS):
    count = video_counts[i]
    total_videos += count
    print(f"{i}: {motion_label:20s} - {count:4d} videos")
print(f"\nTotal: {total_videos} videos")
```

## 5. Extract MediaPipe Keypoints from Videos

```python
KEYPOINTS_DIR = Path("extracted_keypoints")
KEYPOINTS_DIR.mkdir(exist_ok=True)

MP_POSE = mp.solutions.pose

def extract_keypoints_from_video(video_path, confidence_threshold=0.5):
    """
    Extract MediaPipe 33-keypoint pose from video.
    Returns: (n_frames, 33, 3) array with [x, y, z] per keypoint
    """
    cap = cv2.VideoCapture(str(video_path))
    if not cap.isOpened():
        return None, False
    
    keypoints_list = []
    frame_count = 0
    
    with MP_POSE.Pose(
        min_detection_confidence=confidence_threshold,
        min_tracking_confidence=confidence_threshold,
        model_complexity=1,
    ) as pose:
        while True:
            ret, frame = cap.read()
            if not ret:
                break
            
            rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            rgb.flags.writeable = False
            results = pose.process(rgb)
            rgb.flags.writeable = True
            
            if results.pose_landmarks:
                lm = results.pose_landmarks.landmark
                keypoints = np.array([[p.x, p.y, p.z] for p in lm])  # (33, 3)
                keypoints_list.append(keypoints)
                frame_count += 1
    
    cap.release()
    
    if len(keypoints_list) < 5:  # Need minimum 5 frames
        return None, False
    
    return np.array(keypoints_list), True  # (n_frames, 33, 3)

print("Extracting MediaPipe keypoints...")
print("=" * 60)

extraction_stats = {i: {'total': 0, 'success': 0} for i in range(len(MOTION_LABELS))}

for motion_dir in sorted(SELECTED_MOTIONS_DIR.iterdir()):
    if not motion_dir.is_dir():
        continue
    
    motion_id = int(motion_dir.name.split('_')[0])
    motion_name = motion_dir.name[3:].replace('_', ' ')
    
    output_dir = KEYPOINTS_DIR / motion_dir.name
    output_dir.mkdir(exist_ok=True)
    
    videos = list(motion_dir.glob("*.avi")) + list(motion_dir.glob("*.mp4"))
    extraction_stats[motion_id]['total'] = len(videos)
    
    print(f"\n{motion_id}: {motion_name:25s} ({len(videos)} videos)")
    
    for video_path in tqdm(videos):
        keypoints, success = extract_keypoints_from_video(video_path)
        
        if success:
            # Save as .npy file
            output_file = output_dir / (video_path.stem + ".npy")
            np.save(output_file, keypoints)
            extraction_stats[motion_id]['success'] += 1

print("\n" + "="*60)
print("Extraction Summary:")
print("="*60)
total_success = 0
for i, motion_label in enumerate(MOTION_LABELS):
    stats = extraction_stats[i]
    success_rate = (stats['success'] / max(stats['total'], 1)) * 100
    print(f"{i}: {motion_label:20s} - {stats['success']:3d}/{stats['total']:3d} ({success_rate:5.1f}%)")
    total_success += stats['success']

print(f"\nTotal extracted: {total_success} keypoint sequences")
print(f"Saved to: {KEYPOINTS_DIR}/")
```

## 6. Dataset Overview

```python
import json

# Load extracted keypoints
keypoints_info = {}
total_sequences = 0

print("\nDataset Overview:")
print("=" * 80)
print(f"{'Motion Type':<25} {'Files':<8} {'Frames':<15} {'Shape':<20}")
print("=" * 80)

for motion_dir in sorted(KEYPOINTS_DIR.iterdir()):
    if not motion_dir.is_dir():
        continue
    
    motion_id = int(motion_dir.name.split('_')[0])
    motion_name = motion_dir.name[3:].replace('_', ' ')
    
    npy_files = list(motion_dir.glob("*.npy"))
    
    if len(npy_files) == 0:
        continue
    
    # Load first file to check shape
    sample = np.load(npy_files[0])
    total_frames = sum(np.load(f).shape[0] for f in npy_files)
    
    keypoints_info[motion_id] = {
        'name': motion_name,
        'num_files': len(npy_files),
        'total_frames': total_frames,
        'avg_frames': total_frames // len(npy_files)
    }
    total_sequences += len(npy_files)
    
    print(f"{motion_name:<25} {len(npy_files):<8} {total_frames:<15} {str(sample.shape):<20}")

print("=" * 80)
print(f"Total sequences: {total_sequences}")
print(f"Total frames: {sum(info['total_frames'] for info in keypoints_info.values())}")

# Save dataset info
dataset_config = {
    'motion_labels': MOTION_LABELS,
    'num_keypoints': 33,
    'keypoint_dims': 3,  # x, y, z
    'statistics': keypoints_info,
    'total_sequences': total_sequences
}

with open('dataset_info.json', 'w') as f:
    json.dump(dataset_config, f, indent=2)

print(f"\n✓ Dataset info saved to: dataset_info.json")
print("\n✓ Ready for training! Proceed to: 2_train_and_evaluate.ipynb")
```
